## This notebook demos Python data visualizations on the Dementia dataset


```python
# First, we'll import pandas, a data processing and CSV file I/O library
import pandas as pd

# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)


# Next, we'll load the dementia dataset, which is in the ".." directory
dataa = pd.read_csv("Dementia_data.csv") # the dataset is now a Pandas DataFrame

# Let's see what's in the iris data - Jupyter notebooks print the result of the last thing you do
dataa.head()

# Press shift+enter to execute this cell
```

```python
# Let's see how many classes are there
dataa["Label"].value_counts()
```

```python
# The first way we can plot things is using the .plot extension from Pandas dataframes
# We'll use this to make a scatterplot of the data.
dataa.plot(kind="scatter", x="VERB_ratio", y="NOUN_ratio")
```

```python
# We can also use the seaborn library to make a similar plot
sns.jointplot(x="VERB_ratio", y="NOUN_ratio", data=dataa, size=5)
```

```python
sns.FacetGrid(dataa, hue="Label", size=5) \
   .map(plt.scatter, "VERB_ratio", "NOUN_ratio") \
   .add_legend()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Label", y="MODIFIER_ratio", data=dataa)
```

```python
ax = sns.boxplot(x="Label", y="MODIFIER_ratio", data=dataa)
ax = sns.stripplot(x="Label", y="MODIFIER_ratio", data=dataa, jitter=True, edgecolor="gray")
```

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
sns.violinplot(x="Label", y="MODIFIER_ratio", data=dataa, size=6)
```

```python
sns.FacetGrid(dataa, hue="Label", size=6) \
   .map(sns.kdeplot, "MODIFIER_ratio") \
   .add_legend()
```

```python
sns.pairplot(dataa.drop("message", axis=1), size=2)
```

```python
sns.pairplot(dataa.drop("message", axis=1), size=2, diag_kind="histogram")
```
